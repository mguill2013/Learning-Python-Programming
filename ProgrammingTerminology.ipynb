```python
# The following program is a collection of simple Python programs that I wrote to learn the basics of Python.
# I am using the Python Programming MOOC 2025 course as a reference and guide. 
# This ipynb file contains objectives from "Introduction to Programming Part 2: Programming Terminology". 


# A statement is a part of the program which executes something. It often, but not always, refers to a single command.
# For example, print("Hi!") is a statement. number = 2 is also a statement.
# A statement can also be more complicated:

if name == "Anna":
    print("Hi!")
    number = 2

# In the above program, there are two statements (a print statement and an assignment statement) within a conditional statement.


# A block is a group of consecutive statements that are at the same level in the strucutre of the program. 
# For example, the block of a conditional statement contains those statements which are executed only if the condition is true. 

if age > 17:
    # beginning of the conditional block
    print("You are of age!")
    age = age + 1
    print("You are now one year older...")
    # end of the conditional block

print("This print statement is outside of the conditional block.")

# In Python, blocks are expressed by indentation of all code in the same block by the same amount of whitespace.
# The main block of a Python program must always be at the leftmost edge of the file, without any indentation.


# An expression is a bit of code that results in a determined data type. 
# When the program is executed, the expression is evaluated so that it has a value that can then be used in the program. 

2 + 4 + 3
# The above expression is evaluated to a value of 9. The type of the expression is an integer. The python data type of the expression is int.

"abc" + "de"
# The above expression is evaluated to a value of "abcde". The type of the expression is a string. The python data type of the expression is str.

11 / 2 
# The above expression is evaluated to a value of 5.5. The type of the expression is a float. The python data type of the expression is float.

2 * 5 > 9 
# The above expression is evaluated to a value of True. The type of the expression is a boolean value. The python data type of the expression is bool.

# Because all expressions have a type, they can be assigned to variables:
x = 1 + 2

# Simple expressions can be assembled together to form more complicated expressions, for example with arithmetic operations:
y = 3 * x + x**2

```
